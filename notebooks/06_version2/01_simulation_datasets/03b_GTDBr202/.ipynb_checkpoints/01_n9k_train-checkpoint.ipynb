````r
# reading files
read_comm = function(F){
    df = Fread(F) 
    f = unlist(strsplit(F, '/'))
    f_len = length(f)
    df$sim_rep = f[f_len-1]
    df$abund_dist = f[f_len-2]
    df$richness = f[f_len-3]
    return(df)
}
comm = comm_files %>%
    lapply(read_comm) %>%
    rbindlist(use.names=TRUE)
comm
````

````r
# plotting comm abund distribution
p = comm  %>%
    ggplot(aes(Rank, Perc_rel_abund, group=sim_rep, color=abund_dist)) +
    geom_point(size=0.1) +
    labs(y='% abundance') +
    scale_y_log10() +
    theme_bw() +
    facet_wrap(~ richness, scales='free_x', ncol=1) +
    guides(colour = guide_legend(override.aes = list(size=2)))
p.dims(8,7)
plot(p)
````

````r
# plotting comm abund distribution
p = comm  %>%
    filter.(grepl('mean-10', abund_dist)) %>%
    ggplot(aes(Rank, Perc_rel_abund, group=sim_rep, color=abund_dist)) +
    geom_point(size=0.1) +
    labs(y='% abundance') +
    scale_y_log10() +
    theme_bw() +
    facet_wrap(~ richness, scales='free_x', ncol=1) +
    guides(colour = guide_legend(override.aes = list(size=2)))
p.dims(8,7)
plot(p)
````

## Feature summary

````r
# removing empty files from lists of files
rm_empty = function(L){
    L1 = list()
    for(F in L){
        if(file.size(F) >= 1e3){
            L1[[F]] = F
        } else {
            cat('File size too small:', F, '\n')
        }
    }
    return(L1)
}
````

### Stats

````r
# reading tables
read_feat = function(F){
    #cmd = paste0('gunzip -c ', F, ' | head -n 100')
    #df = Fread(cmd=cmd) %>%
    df = Fread(F) %>%
        select.(assembler, contig, position,
                coverage, num_SNPs, min_insert_size_Match, mean_insert_size_Match, mean_mapq_Match, 
                num_proper_Match, num_orphans_Match, num_discordant_Match, 
                mean_al_score_Match, num_proper_SNP, 
                seq_window_entropy, seq_window_perc_gc, Extensive_misassembly) %>%
        summarize.(coverage = mean(coverage, na.rm=TRUE),
                   contig_length = max(position, na.rm=TRUE),
                   num_SNPs = mean(num_SNPs, na.rm=TRUE),
                   min_insert_size_Match = mean(min_insert_size_Match, na.rm=TRUE),
                   mean_insert_size_Match = mean(mean_insert_size_Match, na.rm=TRUE),
                   mean_mapq_Match = mean(mean_mapq_Match, na.rm=TRUE),
                   num_proper_Match = mean(num_proper_Match, na.rm=TRUE),
                   num_orphans_Match = mean(num_orphans_Match, na.rm=TRUE),
                   num_discordant_Match = mean(num_discordant_Match, na.rm=TRUE),
                   mean_al_score_Match = mean(mean_al_score_Match, na.rm=TRUE),
                   num_proper_SNP = mean(num_proper_SNP, na.rm=TRUE),
                   seq_window_entropy = mean(seq_window_entropy, na.rm=TRUE),
                   seq_window_perc_gc = mean(seq_window_perc_gc, na.rm=TRUE),
                   Extensive_misassembly = mean(Extensive_misassembly, na.rm=TRUE),
                   .by=c(assembler, contig))
    x = unlist(strsplit(F, '/'))
    df$richness = x[length(x)-6]
    df$abund_dist = x[length(x)-5]
    df$sim_rep = x[length(x)-4]
    df$read_length = x[length(x)-3]
    df$seq_depth = x[length(x)-2]
    return(df)
}
df = feat_files %>%
    #head(n=3) %>%
    #lapply(read_feat) %>%
    Q(fun=read_feat, F=., n_jobs=threads, verbose=FALSE) %>%
    rbindlist(use.names=TRUE) 
df
````

#### Summary of contig misassemblies

````r
# feature files
P = file.path(work_dir, 'features')
feat_files = list_files(P, 'features.tsv.gz')
feat_files %>% length
feat_files %>% head(n=3)
````

````r
# removing feature files from unused comm-abund distributions
feat_files = feat_files[!grepl("mean-[15]-sigma-2", feat_files)]
feat_files %>% length
feat_files %>% head(n=3)
````

````r
# adding extra simulations
P = file.path(dirname(work_dir), 'GTDBr202_n9k_train_sigma0.5', 'features')
ff = list_files(P, 'features.tsv.gz')
feat_files = c(feat_files, ff)
feat_files %>% length
feat_files %>% head(n=3)
````

````r
# filtering empty files
feat_files = rm_empty(feat_files)
feat_files %>% length
````

````r
# loading contig misassembly info 
read_feat_mis = function(F){
    df = Fread(F) %>%
        select.(assembler, contig, Extensive_misassembly) %>%
        summarize.(Extensive_misassembly = first(Extensive_misassembly),
                   .by=c(assembler, contig))
    x = unlist(strsplit(F, '/'))
    df$richness = x[length(x)-6]
    df$abund_dist = x[length(x)-5]
    df$sim_rep = x[length(x)-4]
    df$read_length = x[length(x)-3]
    df$seq_depth = x[length(x)-2]
    return(df)
}
doParallel::registerDoParallel(8)
true_mis = feat_files %>%
    #head(n=3) %>%
    plyr::llply(read_feat_mis, .parallel=TRUE) %>%
    rbindlist(use.names=TRUE) 
true_mis
````

````r
# checking the number of parameter combinations
true_mis %>%
    distinct.(assembler, richness, abund_dist, sim_rep, read_length, seq_depth)
````

#### metaMIC contig scores

````r
# contig score files
P = file.path(work_dir, 'metaMIC')
score_files = list_files(P, 'metaMIC_contig_score.txt.gz')
score_files %>% length
score_files %>% head(n=3)
````

````r
# removing feature files from unused comm-abund distributions
score_files = score_files[!grepl("mean-[15]-sigma-2", score_files)]
score_files %>% length
score_files %>% head(n=3)
````

````r
# adding extra simulations
P = file.path(dirname(work_dir), 'GTDBr202_n9k_train_sigma0.5', 'metaMIC')
sf = list_files(P, 'metaMIC_contig_score.txt.gz')
score_files = c(score_files, sf)
score_files %>% length
score_files %>% head(n=3)
````

````r
# filtering
score_files = rm_empty(score_files)
score_files %>% length
````

````r
# loading
read_table = function(F){
    df = Fread(F) 
    x = unlist(strsplit(F, '/'))
    df$richness = x[length(x)-7]
    df$abund_dist = x[length(x)-6]
    df$sim_rep = x[length(x)-5]
    df$read_length = x[length(x)-4]
    df$seq_depth = x[length(x)-3]
    df$assembler = x[length(x)-2]
    return(df)
}
doParallel::registerDoParallel(8)
mMIC = score_files %>%
    plyr::llply(read_table, .parallel=TRUE) %>%
    rbindlist(use.names=TRUE) 
mMIC
````

````r
# joining
overlap(mMIC$contig, true_mis$contig)
mMIC = mMIC %>%
    right_join.(true_mis, c('contig', 'assembler', 'richness',
                            'abund_dist','sim_rep','read_length',
                            'seq_depth')) %>%
    mutate.(metaMIC_contig_score = ifelse(is.na(metaMIC_contig_score), 0, 
                                          metaMIC_contig_score)) %>%
    filter.(sim_rep == 1)
mMIC
````

#### Accuracy

````r
# contig cutoff
ms_cutoff = mMIC %>%
    filter.(assembler == 'metaspades') %>%
    .$metaMIC_contig_score %>% quantile(0.95)
cat('metaspades cutoff:', ms_cutoff, '\n')
````

````r
mMIC_s = mMIC %>%
    mutate.(prediction = ifelse(assembler == 'megahit', 
                                 metaMIC_contig_score >= 0.8,
                                 metaMIC_contig_score >= ms_cutoff)) %>%
    mutate.(prediction = ifelse(prediction == TRUE, 1, 0)) %>%
    mutate.(TP = prediction == 1 & Extensive_misassembly == 1,
            TN = prediction == 0 & Extensive_misassembly == 0,
            FP = prediction == 1 & Extensive_misassembly == 0,
            FN = prediction == 0 & Extensive_misassembly == 1) %>%
    summarize.(TP = sum(TP),
               TN = sum(TN),
               FP = sum(FP),
               FN = sum(FN),
               .by=c(assembler, richness, abund_dist, 
                     sim_rep, read_length, seq_depth)) 
mMIC_s
````

````r
# plotting
p = mMIC_s %>%
    filter.(sim_rep == 1) %>%
    mutate.(seq_depth = seq_depth %>% as.Num) %>%
    pivot_longer.(cols=c(-assembler, -richness, -abund_dist, 
                         -sim_rep, -read_length, -seq_depth)) %>%
    ggplot(aes(assembler, value)) +
    geom_boxplot() +
    facet_grid(name ~ seq_depth, scales='free_y') +
    theme_bw()
p.dims(6,4)
plot(p)
````

````r
# function to calc aucpr
calc_aucpr = function(df){
    if(nrow(df) == 0){
        df = data.frame(aucpr = NA,
                        assembler = df$assembler[1],
                        richness = df$richness[1],
                        abund_dist = df$abund_dist[1],
                        sim_rep = df$sim_rep[1],
                        read_length = df$read_length[1],
                        seq_depth = df$seq_depth[1])
        return(df)
    }
    fg = df %>%
        filter.(Extensive_misassembly == 1) %>%
        .$metaMIC_contig_score
    bg = df %>%
        filter.(Extensive_misassembly == 0) %>%
        .$metaMIC_contig_score
    pr = pr.curve(scores.class0 = fg, scores.class1 = bg)
    df = data.frame(aucpr = pr$auc.integral,
                    assembler = df$assembler[1],
                    richness = df$richness[1],
                    abund_dist = df$abund_dist[1],
                    sim_rep = df$sim_rep[1],
                    read_length = df$read_length[1],
                    seq_depth = df$seq_depth[1])
    return(df)
}
#calc_aucpr(mMIC)
````

````r
# per condition
mMIC_s = mMIC %>%
    group_split.(assembler, richness, abund_dist, sim_rep, read_length, seq_depth) %>%
    lapply(calc_aucpr) %>%
    rbindlist(use.names=TRUE)
mMIC_s
````

````r
# checking counts
mMIC_s %>%
    filter.(abund_dist == 'mean-10-sigma-1') %>%
    summarize.(n = n.(),
               .by=c(assembler, richness, read_length, seq_depth))
````

````r
# outliers
mMIC_s %>% filter.(aucpr > 0.3)
````

````r
# plotting
p = mMIC_s %>%
    filter.(sim_rep == 1) %>%
    mutate.(richness = richness %>% as.Num,
            seq_depth = seq_depth %>% as.Num) %>%
    ggplot(aes(assembler, aucpr, color=read_length, shape=abund_dist)) +
    geom_point(size=3, alpha=0.5) +
    facet_grid(seq_depth ~ richness) +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )
p.dims(7,4)
plot(p)
````
